```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Federal Student Loan Portfolio Summary (https://studentaid.gov/data-center/student/portfolio) 
FSA= pd.read_excel('PortfolioSummary.xls', skiprows=5)
```

```python
#renaming columns
FSA = FSA.rename(columns = {'Federal Fiscal Year2': 'Fiscal Year', 
                                          'Unnamed: 1':'Quarter',
                                          'Dollars Outstanding           (in billions)': 'Direct Loans ($ in billions)',
                                          'Recipients3     (in millions)':'Direct Loan Reicipients (in millions)',
                                          'Dollars Outstanding            (in billions)': 'FFEL Loans ($ in billions)',
                                          'Recipients     (in millions)': 'FFEL Loan Recipients (in millions)',
                                          'Dollars Outstanding           (in billions).1' : 'Perkins Loans ($ in billions)',
                                          'Recipients     (in millions).1': 'Perkins Loan Recipients (in millions)',
                                          'Dollars Outstanding            (in billions).1' : 'Total (in billions)',
                                          'Unduplicated Recipients    (in millions)': 'Total Recipients (unduplicated in millions)'
                                         })
```

```python
#dropping missing values
FSA = FSA.dropna(subset = 'Direct Loans ($ in billions)')

#filling missing values
FSA['Fiscal Year'] = FSA['Fiscal Year'].ffill(limit=3)
FSA.head(10)
```

```python
# SLIDE 2 FEDERAL STUDENT LOAN  
# Visualizing FFEL and Direct Loan outstanding loans (note on July 1st 2010 FFEL was discontinued) 
a4_dims = (8.7, 6)
fig, ax = plt.subplots(figsize=a4_dims)
ax0 =ax.set(title='Federal Student Loans')
ax1 = sns.lineplot(x='Fiscal Year', y='Direct Loans ($ in billions)', data=FSA, ci=None)
ax2 = sns.lineplot(x='Fiscal Year', y='FFEL Loans ($ in billions)', data=FSA, ci=None)
ax3 = sns.lineplot(x='Fiscal Year', y= 'Perkins Loans ($ in billions)',data=FSA, ci=None)

plt.plot([2010], [510], 'x', color='black', markeredgewidth=2, markersize=10)

legend = plt.legend(labels=['Direct Loans', 'FFEL Loans', 'Perkins Loans']).set(title='Federal Loans')
```

```python
# Federal Student Aid Portfolio by Loan Type (https://studentaid.gov/data-center/student/portfolio)
FSAloantype = pd.read_excel('PortfoliobyLoanType.xls', skiprows=3)

FSAloantype = FSAloantype.rename(columns={'Unnamed: 0':'Fiscal Year',
                                          'Unnamed: 1': 'Quarter',
                                          'Unnamed: 2': 'Stafford Sub Outstanding (billions)',
                                          'Unnamed: 3': 'Stafford Sub Recipients (millions)',
                                          'Unnamed: 4': 'Stafford Unsub Outstanding (billions)', 
                                          'Unnamed: 5': 'Stafford Unsub Recipients (millions)',
                                          'Unnamed: 6': 'Stafford (Combined) Outstanding (billions)',
                                          'Unnamed: 7': 'Stafford (combined) Recipients (millions)',
                                          'Unnamed: 8': 'Grad PLUS Outstanding (billions)',
                                          'Unnamed: 9': 'Grad PLUS Recipients (millions)',
                                          'Unnamed: 10': 'Parent PLUS Outstanding (billions)',
                                          'Unnamed: 11': 'Parent PLUS Recipients (millions)',
                                          'Unnamed: 12': 'Perkins Outstanding (billions)',
                                          'Unnamed: 13': 'Perkins Recipients (millions)',
                                          'Unnamed: 14': 'Consolidation Outstanding (billions)',
                                          'Unnamed: 15': 'Consolidation Recipients (millions)'                                
                                         })

FSAloantype = FSAloantype.drop(FSAloantype.head(3).index)
FSAloantype = FSAloantype.drop(FSAloantype.tail(5).index)
```

```python
FSAloantype['PLUS loans'] = FSAloantype['Grad PLUS Outstanding (billions)'] + FSAloantype['Parent PLUS Outstanding (billions)']
```

```python
FSAloantype.head(5)
```

```python
a4_dims = (8.7, 6)
fig, ax = plt.subplots(figsize=a4_dims)
ax0 =ax.set(title='Direct Loans')

ax1 = sns.lineplot(x= 'Fiscal Year', y='Stafford Sub Outstanding (billions)', data=FSAloantype, ci=None)
ax2 = sns.lineplot(x= 'Fiscal Year', y='Stafford Unsub Outstanding (billions)', data=FSAloantype, ci=None)
ax3 = sns.lineplot(x= 'Fiscal Year', y='Consolidation Outstanding (billions)', data=FSAloantype, ci=None)
ax4 = sns.lineplot(x= 'Fiscal Year', y='PLUS loans', data=FSAloantype, ci=None)

ax.legend(labels=['Direct Sub Loans', 'Direct Unsub Loans', 'Consolidated Loans', 'PLUS Loans'])
```
