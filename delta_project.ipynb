```python
import pandas as pd
import numpy as np
```

```python
#Importing Delta_Data_Dictionary_1987_2012
#https://nces.ed.gov/ipeds/use-the-data/delta-cost-project-finance-data
delta = pd.read_excel('Delta_Data_Dictionary_1987_2012.xls')
delta
```

```python
#retrieving old data from the delta project
ipeds1 = pd.read_csv('delta_public_00_12.csv', encoding = 'ISO-8859-1', low_memory='False')
ipeds2 = pd.read_csv('delta_public_87_99.csv', encoding='ISO-8859-1', low_memory='False')
```

```python
# selecting relevant IPEDS variables 
variables = ['instname', 'academicyear', 'state', 'control', 'fte_count', 'tuition01',
            'tuition02', 'net_student_tuition', 'federal03', 'state03', 'investment01', 'appliedaid01',
             'grant06', 'grant07', 'institutional_grant_aid_share', 'tuition_discount', 'inst_grant_pct',
             'inst_grant_avg_amount', 'loan_avg_amount', 'associatedegrees', 'bachelordegrees', 'masterdegrees',
             'doctordegrees', 'grad_rate_150_p', 'ftretention_rate', 'year_total_undergrad', 'admssn', 'applcn',
             'total_enrollment'          
            ]

# data definitions of variables
a = delta.loc[delta['Variable'].isin(variables)]
a
```

```python
# extracting and combining delta project variables
ipeds1 = ipeds1[variables]
ipeds2 = ipeds2[variables]
final_delta = pd.concat([ipeds1, ipeds2])
final_delta.head()
```

```python
#removing non public and private schools
final_delta = final_delta.loc[(final_delta['control'] != 3) & (final_delta['control'] != -3)]
control = {2:'private', 1:'public'} 
final_delta = final_delta.replace({'control':control})

final_delta['control'].value_counts()
```

```python
a = final_delta.groupby(['academicyear','state','control']).agg(np.mean)
a
```

```python
#aggregating variables by mean by year state and public/private distinction
b = final_delta.groupby(['academicyear','state','control']).mean()
b = b.reset_index()
b
```

```python
b.to_csv('final_delta.csv')
```
