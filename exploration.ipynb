```python
# Federal Student Loan Portfolio Summary (https://studentaid.gov/data-center/student/portfolio) 
FSAsummary = pd.read_excel('PortfolioSummary.xls', skiprows=5)
FSAsummary = FSAsummary.rename(columns = {'Federal Fiscal Year2': 'Fiscal Year', 
                                          'Unnamed: 1':'Quarter',
                                          'Dollars Outstanding           (in billions)': 'Direct Loans ($ in billions)',
                                          'Recipients3     (in millions)':'Direct Loan Reicipients (in millions)',
                                          'Dollars Outstanding            (in billions)': 'FFEL Loans ($ in billions)',
                                          'Recipients     (in millions)': 'FFEL Loan Recipients (in millions)',
                                          'Dollars Outstanding           (in billions).1' : 'Perkins Loans ($ in billions)',
                                          'Recipients     (in millions).1': 'Perkins Loan Recipients (in millions)',
                                          'Dollars Outstanding            (in billions).1' : 'Total (in billions)',
                                          'Unduplicated Recipients    (in millions)': 'Total Recipients (unduplicated in millions)'
                                         })

FSAsummary = FSAsummary.ffill(limit=3)
FSAsummary = FSAsummary.drop([44,45,46,47], axis=0)

FSAsummary.head(10)
```

```python
# visualization of direct loans per year
sns.lineplot(data=FSAsummary, x='Fiscal Year', y='Direct Loans ($ in billions)')
```

```python
# direct loan summary including direct loans per recipient 
FSAsummaryDL = FSAsummary[['Fiscal Year', 'Quarter', 'Direct Loans ($ in billions)','Direct Loan Reicipients (in millions)']].copy()
FSAsummaryDL['DL/per_r'] = FSAsummaryDL['Direct Loans ($ in billions)']/ FSAsummaryDL['Direct Loan Reicipients (in millions)']
FSAsummaryDL['%change'] = FSAsummaryDL['DL/per_r'].pct_change()
FSAsummaryDL['%change'] = round(FSAsummaryDL['%change'] *100,2)
FSAsummaryDL.tail(10)
```

```python
# visualization of direct loans per year per recipient 
sns.lineplot(data=FSAsummaryDL, x='Fiscal Year', y='DL/per_r')
```

```python
# percent change per year for amount of direct loans  per recipient 
sns.lineplot(data=FSAsummaryDL, x='Fiscal Year', y='%change')
```
